# Grabbing the url links for each whiskey listed in the new-releases page

```python
ua = {"User-Agent":"Mozilla/5.0"}
url = "https://www.whiskybase.com/whiskies/new-releases"
page = requests.get(url, headers=ua)
soup = BeautifulSoup(page.text, "html")
atags = soup.find_all('a', {'class' : 'clickable'})
links = [atag['href'] for atag in atags]
```

# Final function which consolidates all of the test functions from above to grab the whiskey url, whiskey name, user name, user rating, and price

```python
list_of_dicts = []
# links are the specific whiskey pages from the cell above
# creating a global variable list_of_dicts to fill with the information from the function below
def get_ratings(links):
    # Using selenium to log into the site with provided username/password
    url = "https://www.whiskybase.com/account/login?return=https%3A%2F%2Fwww.whiskybase.com%2Fwhiskies%2Ftop1000"
    driver = webdriver.Chrome('/Users/hiowatah/Downloads/chromedriver 5')
    driver.get(url)
    element = driver.find_element_by_name("username")
    element.click()
    element.send_keys("whohuhwhat")
    element = driver.find_element_by_name("password")
    element.click()
    element.send_keys("e&a97v0SXihh")
    element.send_keys(Keys.ENTER)
    # Looping through all of the whiskeys from the list of links provided in the parameter
    for i in links:
        url = i
        driver.get(url)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # Getting the whiskey prices
        # some whiskeys don't have prices listed so try function to bring in prices, otherwise a 'N/A' to 
        # make it easier to clean up once in a DataFrame
        try:
            price = soup.find('div', class_="block-price").find_all('p')[-1].get_text()
        except:
            price = 'N/A'
        
        # Getting the usernames
        users = soup.find_all('div', class_="wb--community-block")[1].find_all('th' ,scope='row')
    
        user_list = []
        for user in users:
            # Was getting some errors on a few pages which didn't have reviews, 
            # so try and except to just grab the information I could
            try:
                user_list.append(user.get_text().strip('\n'))
            except:
                pass
        
        # Getting the ratings 
        ratings = soup.find_all('div', class_="wb--community-block")[1].find_all('div' ,class_="wb--rating")
        ratings_list = []
        for rating in ratings:
            # Was getting some errors on a few pages which didn't have reviews, 
            # so try and except to just grab the information I could
            try:
                ratings_list.append(rating.get_text().strip('\n').split('\n')[0])
            except:
                pass
        # All the lists should be the same length, so consolidating all of the information grabbed above
        # to an empty dictionary and then appending it to the global variable list_of_dicts
        for i in range(len(ratings_list)):
            try:
                whiskey_dict = {}
                whiskey_dict['whiskey'] = url.split('/')[-1]
                whiskey_dict['user'] = user_list[i]
                whiskey_dict['rating'] = ratings_list[i]
                whiskey_dict['price'] = price
                whiskey_dict['url'] = url
                list_of_dicts.append(whiskey_dict)
            except:
                pass
    
    return list_of_dicts
```
